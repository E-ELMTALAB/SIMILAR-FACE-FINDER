```python
### colab face represent extractor 2
### colab feature vectore extractor 

import os
from deepface import DeepFace
import pickle

# Set the path to the parent directory of the nested folders
# parent_dir = "/content/all-these-people-dont-exist/ThisPersonDoesNotExist/ThisPersonDoesNotExist"
parent_dir = "/content/images"
# parent_dir = r"C:\python\COMPUTER VISION\open_cv\object_detection\face_recognition\celeb_look_ailke\face_dataset"
human_feature = []
woman_feature = []
i = 0 
k = 0

# Use os.walk to traverse through all the subdirectories
for subdir, dirs, files in os.walk(parent_dir):
    for file in files:
        # Check if the file is an image (you can modify the condition based on your file extensions)
        if file.endswith(".jpg") or file.endswith(".png") or file.endswith(".jpeg"):
            # Get the full path of the image
            img_path = os.path.join(subdir, file)
            

            # gender = DeepFace.analyze(img_path=img_path , actions=["gender"])[0]["dominant_gender"]
            
            i = i + 1
            print(i)
            try:
              feature_vec = DeepFace.represent(img_path=img_path)
              feature_dict = {"image_path" : img_path , "feature_vector" : feature_vec}
              human_feature.append(feature_dict)
              k = k + 1
            except:
              pass

            if k == 5000:

                print("step : " + str(i) + " , writting the feautes to a pickel file...")
                with open(f'HUMAN{i}.pickle', 'wb') as file:
        
                    # A new file will be created
                    pickle.dump(human_feature, file)

                k = 0
                human_feature.clear()



# print(len(man_feature))

with open('HUMAN_last.pickle', 'wb') as file:
      
    # A new file will be created
    pickle.dump(human_feature, file)

# with open('WOMAN.pickle', 'wb') as file:
      
#     # A new file will be created
#     pickle.dump(woman_feature, file)

```
